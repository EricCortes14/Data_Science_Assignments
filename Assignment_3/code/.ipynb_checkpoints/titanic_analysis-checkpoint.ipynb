We are going to study the survival rate of passengers on titanic and what variables affected survival.

Load the dataset in `titanic.xls`. It contains data on all the passengers that travelled on the Titanic.

```python
from IPython.core.display import HTML
import pandas as pd
```

```python
HTML(filename='../data/titanic.html')
```

```python
# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=None)
t_df.head()
```

```python
#obtain the column names from row 0
columns = t_df.iloc[0]

#rename columns to the values of row 0
t_df.columns = columns

#drop row 0
t_df = t_df.drop(t_df.index[0])

#convert age and survived columns to float
t_df['survived'] = t_df['survived'].astype(float)
t_df['age'] = t_df['age'].astype(float)

#display dataset
t_df
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (10 pts)*** 

```python
#make a temporary dataset from the original
tmp = t_df

#groupby sex
tmp = tmp.groupby('sex')

#aggregate data to find the number of survivors and number of each sex on board
tmp = tmp['survived'].agg(['sum','count']).reset_index()

#calculate the proportion of who survived by race as a percent
tmp['proportion as percent'] = (tmp['sum']/tmp['count'])*100

#rename columns
tmp.columns = ['sex', 'survived', 'total', 'proportion (as percent)']

#display temporary dataset
tmp
```

Conclusion:

As can be observed the percent of females that survived the titanic accident was significantly much more higher then the percent of males who survived. The proportions indicate that the proportion of females that survived was more then 3 times that of the proportion of males that survived. Thus, this indicates that sex did have a significant impact on the survival of the passengers.

*** 2. Calculate the same proportion, but by class and sex. (10 pts)*** 

```python
#make a temporary dataset of original dataset
tmp2 = t_df

#group by class and sex
tmp2 = tmp2.groupby(['pclass', 'sex'])

#aggregate data to obtain the number of people who survived and the total number of people per category
tmp2 = tmp2['survived'].agg(['sum','count']).reset_index()

#calculate proportion of survival as a percent
tmp2['proportion (normalized by 100 people)'] = (tmp2['sum']/tmp2['count'])*100

#rename columns
tmp2.columns = ['pclass', 'sex', 'survived', 'total', 'proportion (as percent)']

#display dataset
tmp2
```

Conclusion:

Observing the proportions once again we notice that under each class females had a higher proportion of survival. Now including the class variable it can also be noted that the proportion of survial decreased going down the classes for the female proportions. The males also decreased in proportion from the first class to the second and third but both second class males and third class males had almost the same survival proportion. This means that class did have an impact on survial for females and also had some impact on the survival proporiton for males but not as significant. Overall class had an impact on survial but not as impactful as sex had.

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (20 pts)***

```python
#make a copy of the original dataset and store as temporary dataset
tmp3 = t_df

#create bins to categorize each person per age group
bins= [0, 14, 21, 65, 130]

#create labels that each person will be labeled as according to their age group
labels = ['under 14','14-20','21-64','65+']

#create a new column and do the apporpriate labeling
tmp3['ageGroup'] = pd.cut(tmp3['age'], bins=bins, labels=labels, right=False)

#keep only the needed columns
tmp3 = tmp3[['ageGroup', 'pclass', 'sex', 'survived']]

#group by ageGroup, class and sex
tmp3 = tmp3.groupby(['ageGroup', 'pclass', 'sex'])

#aggregate to calculate the number who survived per category and the total number of people per category
tmp3 = tmp3['survived'].agg(['sum','count']).reset_index()

#calculate proportion as percent
tmp3['proportion (as percent)'] = (tmp3['sum']/tmp3['count'])*100

#rename columns
tmp3.columns = ['ageGroup', 'pclass', 'sex', 'survived', 'total', 'proportion (as percent)']

#display dataset
tmp3
```

Conclusion:
    
Age group also had an impact on the survival proportions of the passengers on the titanic. Observing the proportions of the passengers that were less then 14 years of age one can notice that their proportions of survival for both first and second class were 100 percent. From there going up the age group the proportions start to decrease once again until reaching the age group of 65+ where most proportions were 0 percent except for some outliars. Age group was another factor that contributed to the survial proportions of the passengers
